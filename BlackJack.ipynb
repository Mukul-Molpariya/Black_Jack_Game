```python
import random

suits = ('Heart','Diamond','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten',
        'King','Queen','Jack','Ace',)
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,
        'King':10,'Queen':10,'Jack':10,'Ace':11,}
playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
        
```

```python
class Deck:
    
    def __init__(self):
        self.deck = [] #start with on empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n' + card.__str__()
        return 'The deck has: ' + deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
        
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
        
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []   # start with an empty list as we did in the Deck class
        self.value = 0    # start with zero value
        self.aces = 0     # add on attribute to keep track of aces
        
    def add_card(self,card):
        # card passed in
        # from deck.deal() --> single card(suit,rank)
        self.cards.append(card)
        self.value += values[card.rank]
        
        # track aces
        
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_ace(self):
        
        # If total value >21 AND i still have an ace
        # Than change my ace to be a 1 instead of n 11
        while self.value > 21 and self.aces > 0:
            self.value -= 10
            self.aces -= 1
        
```

```python
class Chips:
    
    def __init__(self, total = 100):
        self.total = total # This can be set to a default value or supplies by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips would you like to bet : "))
        except ValueErr:
            print("Sorry please provide an integer")
        else:
            if chips.bet > chips.total:
                print("Sorry , you do not have enough chips! You have: {}".format(chips.total))
            else:
                break
```

```python
def hit(deck,hand):
    
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing # to control an upcoming while loop
    
    while True:
        x = input("Hit or Stand ? Enter h or s") # HIT # hh # stand
        
        if x[0].lower() == 'h':
            hit(deck,hand)
        
        elif x[0].lower() == 's':
            print("Player Stand Dealer's turn")
            playing = False
            
        else:
            print("Sorry , I did no understand that , please enter h or s only!")
            continue
        break
```

```python
def show_some(player,dealer):
    
    # Show only ONE of the dealer's cards
    
    print("\n Dealer's hand : ")
    print("First card hidden!")
    print(dealer.cards[1])
    
    # Show all (2 cards) of the player's hands/cards
    
    print("\n Player's hand : ")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    
    # Show all the dealer's cards
    print("\n Dealer's hand : ")
    for card in dealer.cards:
        print(card)
    # Calculae and display value (J+K == 20)
    print(f"Value of Dealer'shand is : {dealer.value} ")
    
    # Show all the players card
    
    print("\n Player's hand : ")
    for card in player.cards:
        print(card)
    print(f"Value of Player'shand is : {player.value} ")
```
